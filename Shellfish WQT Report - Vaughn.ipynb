### URL Queries
* Base Path:   
  * https://waterquality.piercecountywa.org/KiWIS/KiWIS?datasource=0&service=kisters&type=queryServices&request=
* Non-timeseries parameters as timeseries:  
  * getWqmSampleValuesAsTimeseries&period=*[Period]*&ts_path=*[Timeseries Path]*&format=csv&metadata=true
  * period can be form 'from + to', 'from', or 'PxYxMxD' (i.e. P36M for last 36 month). 
  * ts_path should be in the form: site_no/station_no/parametertype_shortname
  * ex:  
    getWqmSampleValuesAsTimeseries&period=P36M&ts_path=15/KP_VAUG_0.16/FecalColiform_MF&format=csv&&csvdiv=,&metadata=true
* Timeseries for rainfall data:
  * 'getTimeseriesValues&'+*[Period]*+'&ts_id='+*[ts_id]*'&&format=csv&csvdiv=,&metadata=true'


```python
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib as mpl
%matplotlib inline

# enter time period:
# for last 3 years, use 'period=P36M' (past 36 months) or 'period=P3Y' (past 3 years)
# or, for specific range (from x to y), use 'from=yyyy-mm-dd&to=yyyy-mm-dd'
# alternately, omit 'to=*' to to default to today's date ('from=yyyy-mm-dd')
tmp = 'period=P36M'

# enter sampling location (i.e. ts_path elements)
site_no = '15'
station_no = 'KP_VAUG_0.16'
parametertype_shortname = 'FecalColiform_MF'

# enter weather station data
ts_id = '25555010'

# set up url query for request - returns csv, but need to set csvdiv to ',' (defaults ';')
ts_path = site_no + '/' + station_no + '/' + parametertype_shortname
param_url = 'https://waterquality.piercecountywa.org/KiWIS/KiWIS?datasource=0&service=kisters&type=queryServices&request=getWqmSampleValuesAsTimeseries&'+tmp+'&ts_path='+ts_path+'&format=csv&csvdiv=,&metadata=true'

# set up url query for rainfall (timeseries)
rain_url = 'https://waterquality.piercecountywa.org/KiWIS/KiWIS?datasource=0&service=kisters&type=queryServices&request=getTimeseriesValues&'+tmp+'&ts_id='+ts_id+'&&format=csv&csvdiv=,&metadata=true'

# write datasets to pandas dataframes
bdf = pd.read_csv(param_url, skiprows=9)
rdf = pd.read_csv(rain_url, skiprows=11).fillna(0)

# Format column header strings for clarity and consistency
tsp = 'Timestamp'
fecs = 'Fecal Coliform (MF)'
rain = '24hr Rainfall (in)'

bdf.columns = [tsp, fecs]
rdf.columns = [tsp, rain]

# convert timestamp column values from string to datetime
rdf['Timestamp'] = pd.to_datetime(rdf[tsp])
bdf['Timestamp'] = pd.to_datetime(bdf[tsp])

# filter out nondetects to simplify graph makes unusable for statistical analysis 
bdf1=bdf[bdf[fecs]!=1]


#set up  plot figure environment:
fig, ax1 = plt.subplots(figsize=(25,10))
darkgray = '#6d6d6d'

# twinx() method allows us to stack the graphs
ax2 = ax1.twinx()

# set plot variables and customizations
ax2.plot(rdf[tsp], rdf[rain], linewidth = 1, color=darkgray)

ax2.set_ylabel('24-Hour Rainfall Total (inches)', fontsize=18, color=darkgray, labelpad=12)
ax2.set_ylim(-0.01, 2.5)
ax2.tick_params(colors=darkgray, labelsize=18, axis='both', direction='inout')
ax2.patch.set_visible(True)
ax2.legend(loc=1, fontsize=18, frameon=True, facecolor='white', edgecolor='black')

ax1.plot_date(bdf1[tsp], bdf1[fecs], color='red', mec='black', ms=10)
# add reference lines for regulatory context
ax1.axhline(y=43, ls='--', label = 'DOH Shellfhish Limit (90th %ile <= 43)').set_zorder(1)
ax1.axhline(y=100, ls='--', color='#827EBC', label = 'Ecology Recreation Limit (90th %ile <= 43)').set_zorder(1)
# define remaining plot settings
ax1.set_ylabel('Fecal Coliform (CFU/100mL)', fontsize=18, color=darkgray, labelpad=12)
ax1.set_ylim(-20, 2000)
ax1.tick_params(colors=darkgray, labelsize=18)
ax1.set_zorder(5)
ax1.patch.set_visible(False)
ax1.set_xlabel('Date', fontsize=24, color=darkgray, visible=True, labelpad=12)
ax1.legend(loc=2, fontsize=18, frameon=True, facecolor='white', edgecolor='black')
# draw dashed line at 43 cfu for reference (DOH action level)


# save the image as a png using the most recent sample date as ID
output = station_no+'_'+str(bdf[tsp].iloc[-1])
plt.savefig(output[0:20] +'.png', bbox_inches='tight', dpi=400)

```

```python
# use merge_asof function to join dataframes on nearest date match
# using the left (bacteria dataframe/bdf) as the join key
df_merge = pd.merge_asof(bdf1, rdf, on='Timestamp').dropna()
df_merge['Timestamp'] = df_merge['Timestamp'].dt.strftime('%m-%d-%Y')

# save to file
df_merge.to_csv(output[0:20] +'.csv')
```
